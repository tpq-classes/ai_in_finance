#### Activation

```python
def sigmoid(x, deriv=False):
    if deriv:
        return sigmoid(x) * (1 - sigmoid(x))
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, np.where(x > 0, 1, 0), 'y--', label='step function')
plt.plot(x, sigmoid(x), 'r', label='sigmoid')
plt.plot(x, sigmoid(x, True), '--', label='derivative')
plt.legend();
```

#### Steps

```python
features = 5
samples = 4
units = 5
```

```python
l0 = np.random.random((samples, features))
w0 = np.random.random((features, units))  * 0.15
w1 = np.random.random((units, 1)) * 0.15
y = np.random.random((samples, 1))
```

```python
a = 0.01
steps = 20000
```

```python
for s in range(1, steps + 1):
    l1 = np.dot(l0, w0)
    l2 = np.dot(l1, w1)
    e2 = l2 - y
    u2 = a * np.dot(l1.T, e2)
    w1 -= u2
    e1 = np.dot(e2, w1.T)
    u1 = a * np.dot(l0.T, e1)
    w0 -= u1
    mse = (e2 ** 2).mean()
    if s % 2000 == 0:
        print(f'step={s:5d} | mse={mse:.5f}')
```

#### Steps

```python
features = 5
samples = 10
units = 10
```

```python
np.random.seed(200)
l0 = np.random.randint(0, 2, (samples, features))
w0 = np.random.random((features, units))
w1 = np.random.random((units, 1))
y = np.random.randint(0, 2, (samples, 1))
```

```python
a = 0.1
steps = 20000
```

```python
for s in range(1, steps + 1):
    l1 = sigmoid(np.dot(l0, w0))
    l2 = sigmoid(np.dot(l1, w1))
    e2 = l2 - y
    d2 = e2 * sigmoid(l2, True)
    u2 = a * np.dot(l1.T, d2)
    w1 -= u2
    e1 = np.dot(d2, w1.T)
    d1 = e1 * sigmoid(l1, True)
    u1 = a * np.dot(l0.T, d1)
    w0 -= u1
    mse = (e2 ** 2).mean()
    if s % 2000 == 0:
        print(f'step={s:5d} | mse={mse:.5f}')
```

```python
acc = l2.round() == y
acc
```

```python
sum(acc) / len(acc)
```

## Neural Networks with Keras

### Estimation

```python
import numpy as np
```

```python
x = np.linspace(0, 3 * np.pi, 50)
y = np.cos(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro');
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
```

```python
import tensorflow as tf
from keras.layers import Dense
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(4 * 48, input_dim=1, activation='relu'))
model.add(Dense(4 * 24, activation='relu'))
```

```python
# estimation
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='adam', metrics=['mse', 'accuracy'])
```

```python
model.summary()
```

```python
%time model.fit(x, y, epochs=1000, verbose=False)
```

```python
scores = model.evaluate(x, y)
```

```python
print('mse: %.5f' % (scores[1]))
```

```python
# estimation
pred = model.predict(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro', label='original data')
plt.plot(x, pred, label='prediction')
plt.legend();
```

### Binary Classification

```python
yc = np.where(y > 0, 1, 0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'm^')
plt.plot(x, yc, 'ro');
```

```python
model = Sequential()
model.add(Dense(48, input_dim=1, activation='relu'))
model.add(Dense(24, activation='relu'))
```

```python
# classification
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
%time model.fit(x, yc, epochs=1500, verbose=False)
```

```python
scores = model.evaluate(x, yc)
```

```python
print('accuracy: %.2f%%' % (scores[1] * 100))
```

```python
# classification
pred = np.where(model.predict(x) > 0.5, 1, 0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, yc, 'ro', label='original data')
plt.plot(x, pred, label='prediction')
plt.legend();
```
